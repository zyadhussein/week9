This code aims to clean the training dataset provided by Data Glacier and Kaggle. the dataset consists of twitter tweets. the dataset has been stored in csv file and some of the tweets had emojis and symbols that csv file didn't store perfectly so the symbos have turned into special charcters and sometimes ascii code or unicode. to tackle this issue and clean the dataset to make it ready for training our model, regex methods were used. the methods were used to remove any special characters, mix of words and numbers, unicode, and any other unidentifiable symbol. hence, providing a clean dataset. the cleaning was done without removing the hashtag symbols as it is believed that hashtags contains emotional information that could help the model in learning and detecting hate speech. it was attempted to remove white space. however, it was found out that removing it can result in dataset deterioration in our case. after the cleaning process implementation, it was found out that the word 'user' was repeated many times without a meaning so word.replace function was used to remove it as shown below. other issues were noticed in the dataset such as mislabelling the datset which is still in progress of modification. such issue could contribte on the decrease of accuracy results. however, it is still in progress of finding a solution to this issue instead of manually modifying it. the dataset also contained some mixed feelings tweets which were labelled into either good or bad. it was thought to remove them and treat them as outliers. however, it is believed that some anomalies could actually help the model trained to be more robust by introducing mixed feelings tweets as seen in the dataset.

loading data using dask as it is fast enough to work with and more organised method besides that it has many functions that can help in tthe manipulation of the dataset.

```python
import dask.dataframe as dd
import re
train=dd.read_csv('train_E6oV3lV.csv')
test=dd.read_csv('test_tweets_anuFYb8.csv')
```

Data cleaning part using regex (regular expression). this parts take a while in implementation process due to the size of the dataset.

```python
train['tweet'].compute()
for i in range(0,31962):

    x=train.loc[i,'tweet'].values
    f=x.compute()
    #this is cleaning a sample text extracted from the dataset using regex as requested.
    #
    cleaning = re.sub(r'[^\w\s\d#]', '', f[0])#remove anything that s not words, numbers or spaces
    #but not hashtags as they contain good information
    cleaning = re.sub(r'@\w+', '', cleaning)# removing @user (optional)
    #cleaning = re.sub(r'\s+', ' ', cleaning).strip()#removing whitespaces
    cleaning = re.sub(r'http\S+|www\S+|https\S+', '', cleaning)#removing urls
    cleaned = cleaning.encode('ascii', 'ignore').decode('ascii')#remove unicode characters
    data.append(cleaned)
```
